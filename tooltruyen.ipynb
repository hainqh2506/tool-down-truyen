```python
import requests
from bs4 import BeautifulSoup
import os

def crawl_chapter(url, title_selector, content_selector, output_dir):
    # Tải nội dung trang web
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')

    # Trích xuất tiêu đề
    title_element = soup.select_one(title_selector)
    title = title_element.text.strip() if title_element else "Không tìm thấy tiêu đề"

    # Trích xuất nội dung
    content_element = soup.select_one(content_selector)
    content = content_element.decode_contents() if content_element else "Không tìm thấy nội dung"

    # Tạo HTML mới chỉ với tiêu đề và nội dung
    new_html = f"""
    <!DOCTYPE html>
    <html lang="vi">
    <head>
        <meta charset="UTF-8">
        <meta name="viewport" content="width=device-width, initial-scale=1.0">
        <title>{title}</title>
    </head>
    <body>
        <h1>{title}</h1>
        <div>{content}</div>
    </body>
    </html>
    """

    # Tạo tên file từ tiêu đề
    filename = f"output.html"  # Giới hạn độ dài tên file
    filepath = os.path.join(output_dir, filename)

    # Lưu file
    with open(filepath, 'w', encoding='utf-8') as f:
        f.write(new_html)

    print(f"Đã lưu file: {filepath}")

# Thông tin crawl
url = "https://truyenfull4.com/legendary-moonlight-sculptor-con-duong-de-vuong/quyen-1-chuong-1.html"
title_selector = "a.chapter-title"
content_selector = "#chapter-c"
output_dir = r"D:\py\truyen crawl"

# Đảm bảo thư mục đầu ra tồn tại
os.makedirs(output_dir, exist_ok=True)

# Thực hiện crawl
crawl_chapter(url, title_selector, content_selector, output_dir)
```

```python
import requests
from bs4 import BeautifulSoup

# URL của trang chứa danh sách các chương
base_url = "https://truyenfull.com/legendary-moonlight-sculptor-con-duong-de-vuong-f3.13140/"
chapter_list = []

# Hàm để lấy tên và link các chương
def get_chapters(url):
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')
    
    chapters = soup.find_all('ul', class_='list-chapter')
    for chapter in chapters:
        links = chapter.find_all('a')
        for link in links:
            chapter_name = link.get_text()
            chapter_url = link.get('href')
            chapter_list.append({'name': chapter_name, 'url': chapter_url})

# Duyệt qua từng trang của danh sách các chương
for page_num in range(1, 14):  # Điều chỉnh phạm vi theo số trang thực tế
    page_url = f"{base_url}trang-{page_num}/#chapter-list"
    get_chapters(page_url)

# Tạo file HTML để lưu kết quả
html_content = """
<!DOCTYPE html>
<html lang="en">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>Chapter List</title>
</head>
<body>
    <h1>Chapter List</h1>
    <ul>
"""

for chapter in chapter_list:
    html_content += f'<li><a href="{chapter["url"]}">{chapter["name"]}</a></li>\n'

html_content += """
    </ul>
</body>
</html>
"""

# Lưu nội dung HTML vào file
with open("chapter_list.html", "w", encoding='utf-8') as file:
    file.write(html_content)

print("Crawl completed and HTML file created.")
```

```python
from bs4 import BeautifulSoup
import os

# Đường dẫn đến file HTML gốc
input_path = r"D:\py\truyen crawl\chapter_list.html"

# Đường dẫn để lưu file HTML mới
output_path = r"D:\py\truyen crawl\chapter_links.html"

# Đọc file HTML gốc
with open(input_path, 'r', encoding='utf-8') as f:
    html_content = f.read()

# Phân tích HTML
soup = BeautifulSoup(html_content, 'html.parser')

# Tìm tất cả các chapter
chapters = soup.select("li.wp-manga-chapter")

# Chuẩn bị dữ liệu để lưu
chapter_links = []
for chapter in chapters:
    link = chapter.find('a')['href']
    title = chapter.find('a').text.strip()
    chapter_links.append(f"<li><a href='{link}'>{title}</a></li>\n")

# Tạo HTML mới chỉ với các link
new_html = f"""
<!DOCTYPE html>
<html lang="vi">
<head>
    <meta charset="UTF-8">
    <title>Danh sách Link Chapter - Con Đường Đế Vương</title>
</head>
<body>
    <h1>Danh sách Link Chapter - Con Đường Đế Vương</h1>
    <ul>
        {"".join(chapter_links)}
    </ul>
</body>
</html>
"""

# Lưu file HTML mới
with open(output_path, 'w', encoding='utf-8') as f:
    f.write(new_html)

print(f"Đã lưu danh sách link chapter vào {output_path}")
```

```python
import os
import requests
from bs4 import BeautifulSoup
import time
import random

# Đường dẫn đến file HTML chứa danh sách link
input_file = r"D:\py\truyen crawl\chapter_links copy.html"

# Đường dẫn để lưu file kết quả
output_file = r"D:\py\truyen crawl\full_story.html"

def get_chapter_links(file_path):
    with open(file_path, 'r', encoding='utf-8') as file:
        soup = BeautifulSoup(file, 'html.parser')
        return [(a['href'], a.text) for a in soup.find_all('a', href=True)]

def crawl_chapter(url):
    headers = {
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'
    }
    try:
        response = requests.get(url, headers=headers, timeout=10)
        response.raise_for_status()
        
        print(f"Status Code: {response.status_code}")
        
        soup = BeautifulSoup(response.text, 'html.parser')
        
        # Sử dụng selector chính xác
        content = soup.select_one("div.reading-content")
        
        if content:
            print("Tìm thấy nội dung")
            
            # Xử lý tiêu đề
            title = content.find('h3')
            if title:
                title_text = title.text.strip()
            else:
                title_text = "Không tìm thấy tiêu đề"
            
            # Xử lý nội dung chapter
            
            chapter_content = content.select_one("div.text-left")
            if chapter_content:
                # Loại bỏ các phần tử không cần thiết
                for element in chapter_content.select("div.chapter-warning, input"):
                    element.extract()
                
                chapter_text = chapter_content.get_text(separator='\n', strip=True)
                return title_text, chapter_text
            else:
                print("Không tìm thấy nội dung trong div.text-left")
        else:
            print("Không tìm thấy div.reading-content")
        
        return "Không tìm thấy tiêu đề", "Không tìm thấy nội dung"
    except requests.RequestException as e:
        print(f"Lỗi khi crawl {url}: {e}")
        return "Lỗi", f"Không thể crawl: {str(e)}"

def main():
    chapter_links = get_chapter_links(input_file)
    
    with open(output_file, 'w', encoding='utf-8') as out_file:
        out_file.write("""<!DOCTYPE html>
<html lang="vi">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>Legendary Moonlight Sculptor - Con Đường Đế Vương</title>
    <link rel="stylesheet" href="styles.css">
</head>
<body>
    <div class="content">
        <h1>Legendary Moonlight Sculptor - Con Đường Đế Vương</h1>
""")
        
        for link, link_text in chapter_links:
            print(f"\nĐang crawl: {link}")
            title, chapter_content = crawl_chapter(link)
            out_file.write(f'        <div class="chapter">\n')
            out_file.write(f'            <h2>{link_text}</h2>\n')
            out_file.write(f'            <h3>{title}</h3>\n')
            out_file.write(f'            <div class="chapter-content">\n')
            out_file.write(f'                <p>{chapter_content}</p>\n')
            out_file.write(f'            </div>\n')
            out_file.write(f'        </div>\n')
            
            time.sleep(random.uniform(1, 3))
        
        out_file.write("""    </div>
</body>
</html>""")

    print(f"\nHoàn thành! Kết quả được lưu tại: {output_file}")

if __name__ == "__main__":
    main()
```

```python
import requests
from bs4 import BeautifulSoup
import os


def crawl_and_save(url, selector, output_path):
    # Gửi request đến trang web
    headers = {
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'
    }
    response = requests.get(url,headers=headers, timeout=10)
    
    # Kiểm tra nếu request thành công
    if response.status_code == 200:
        # Parse nội dung HTML
        soup = BeautifulSoup(response.text, 'html.parser')
        
        # Tìm phần tử theo CSS selector
        content = soup.select_one(selector)
        
        if content:
            # Tạo thư mục nếu chưa tồn tại
            os.makedirs(os.path.dirname(output_path), exist_ok=True)
            
            # Lưu nội dung vào file HTML
            with open(output_path, 'w', encoding='utf-8') as f:
                f.write(str(content))
            print(f"Đã lưu nội dung vào {output_path}")
        else:
            print("Không tìm thấy nội dung với selector đã cho")
    else:
        print(f"Không thể truy cập trang web. Mã trạng thái: {response.status_code}")

# Thông tin crawl
url = "https://truyenz.info/manga/n-con-duong-de-vuong/quyen-51-chuong-6-1/"
selector= "body > div.wrap > div > div.site-content > div > div > div > div > div > div > div.c-blog-post > div.entry-content > div > div > div.reading-content"
output_path = r"D:\py\crawled_content.html"

# Thực hiện crawl và lưu file
crawl_and_save(url, selector, output_path)
```

```python
import requests
from bs4 import BeautifulSoup
import os

# Hàm để crawl và lấy nội dung từ URL
def crawl_and_get_content(url, selector):
    # Gửi request đến trang web
    headers = {
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'
    }
    response = requests.get(url, headers=headers, timeout=10)
    
    # Kiểm tra nếu request thành công
    if response.status_code == 200:
        # Parse nội dung HTML
        soup = BeautifulSoup(response.text, 'html.parser')
        
        # Tìm phần tử theo CSS selector
        content = soup.select_one(selector)
        
        if content:
            return str(content)
        else:
            print(f"Không tìm thấy nội dung với selector đã cho tại {url}")
            return ""
    else:
        print(f"Không thể truy cập trang web. Mã trạng thái: {response.status_code}")
        return ""

# Hàm để lấy URL từ file HTML
def extract_urls_from_file(file_path):
    with open(file_path, 'r', encoding='utf-8') as file:
        content = file.read()
    soup = BeautifulSoup(content, 'html.parser')
    urls = [a['href'] for a in soup.find_all('a', href=True)]
    return urls

# Đường dẫn đến file chứa các link
chapter_links_file = r'D:\py\truyen crawl\chapter_links.html'
urls = extract_urls_from_file(chapter_links_file)

# CSS selector và file đầu ra để lưu các nội dung crawl được
selector = "body > div.wrap > div > div.site-content > div > div > div > div > div > div > div.c-blog-post > div.entry-content > div > div > div.reading-content"
output_file = r"D:\py\truyen crawl\all_chapters.html"

# HTML template
html_template_start = """
<!DOCTYPE html>
<html lang="vi">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>Legendary Moonlight Sculptor - Con Đường Đế Vương</title>
    <link rel="stylesheet" href="style copy.css">
</head>
<body>
"""

html_template_end = """
</body>
</html>
"""

# Ghi nội dung vào file đầu ra
with open(output_file, 'w', encoding='utf-8') as f:
    f.write(html_template_start)
    for idx, url in enumerate(urls, start=1):
        chapter_content = crawl_and_get_content(url, selector)
        if chapter_content:
            f.write(chapter_content)
            print(f"đã crawl chap {idx}")
    f.write(html_template_end)

print(f"Đã lưu tất cả nội dung vào {output_file}")
```
